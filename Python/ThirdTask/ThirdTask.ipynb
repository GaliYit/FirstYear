```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def rotate_matrix_horizontally(matrix):
    matrix = np.flipud(matrix)
    plt.matshow(matrix, cmap='gray')
    
def rotate_matrix_90deg(matrix):
    matrix = np.rot90(matrix)
    plt.matshow(matrix, cmap='gray')
    
def even_row_column(matrix):
    return matrix[::2, ::2]
```

```python
image = np.load('matrix.npy')
bw = np.mean(image, axis=2)
plt.matshow(bw, cmap='gray')
```

```python
rotate_matrix_horizontally(bw)
```

```python
rotate_matrix_90deg(bw)
```

```python
new_image = even_row_column(bw)
print(bw.shape)
print(new_image.shape)
plt.matshow(new_image, cmap='gray')
new_image
```
